*Data_augmentation*

```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import os
from matplotlib import pyplot as plt
import cv2
```

```python
img = load_img('/content/car.jfif')
```

```python
plt.imshow(img)
```

```python
cv2.imread("/content/car.jfif").shape
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        brightness_range=(0.2, 0.3),
        fill_mode='nearest')
```

```python
x = img_to_array(img)  
x = x.reshape((1,) + x.shape)  
```

```python
i = 0
for batch in datagen.flow(x,batch_size=1,save_to_dir='/content/drive/MyDrive/data_augmentation', save_prefix='car', save_format='jpeg'):
    i += 1
    if i >15:
        break  # otherwise the generator would loop indefinitely
```
